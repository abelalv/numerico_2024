# Propagacin del error

La propagacin del error es un concepto fundamental en mtodos numricos y en cualquier tipo de clculo que involucre mediciones y operaciones matemticas. Se refiere a cmo los errores en las mediciones de los datos se propagan a travs de los clculos y afectan el resultado final.

## Ejemplo
Se quiere calcular el rea de un rectngulo con base $b$ y altura $h$. Se miden las dimensiones del rectngulo y se obtienen los valores $b = 2.5 \pm 0.1$ y $h = 3.0 \pm 0.1$. El rea del rectngulo es $A = b \times h$. Cul es el valor de $A$ y cul es su incertidumbre?

Note que aqu se la incertidumbre seria de 0.01. Ahora si calculamos el perimetro del rectangulo, el cual es $P = 2b + 2h$, la incertidumbre seria de 0.4. Esto se debe a que el error se propaga a traves de la suma.

# Mal condicionamiento

Supongamos que queremos resolver el siguiente sistema de ecuaciones lineales:

$$\begin{matrix}7x + 2y = 6 \\ 1x + 0.28y = 2\end{matrix}$$

Note que la solucin exacta del problema es $x=[58,-200]$

+ Comprubelo

+ Ahora si supongamos que el sistema  tiene un pequeo error de medicin, as el sistema correcto es

$$\begin{matrix}7x + 2y = 6 \\ 1x + 0.285y = 2\end{matrix}$$

+ Resuelva el sistema de ecuaciones y compruebe la solucin, Qu observa?



Observamos que un ligero cambio en la matriz de coeficientes del sistema conlleva cambios significativos en la solucin. En este caso particular, las dos soluciones son exactas, ya que solo hay que resolver este problema de manera directa. Sin embargo, si dicha matriz de coeficientes se involucra en un mtodo iterativo y el vector de soluciones, x, debe calcularse en cada iteracin, pequeos errores de redondeo en algunos de los coeficientes pueden amplificarse y propagarse, lo que resultara en una solucin final incorrecta proporcionada por el mtodo.

En situaciones como esta, se clasifica el problema como mal condicionado. Resolver problemas mal condicionados es complejo y, en muchas ocasiones, requiere una reevaluacin del enfoque para evitar depender de trminos que podran verse afectados de manera significativa por los errores de redondeo.

# Estabilidad

Nos interesa crear mtodos numricos que sean estables, es decir, que no amplifiquen los errores de redondeo.
Para analizar la estabilidad de un mtodo numrico, se puede hacer un anlisis de estabilidad, que consiste en analizar el error cometido en cada paso del mtodo numrico y calcular el orden de magnitud del error total cometido.

**Definicin** 
Suponga que $E_0$ denota un error inicial y que $E_n$ denota el error cometido en el paso $n$ del mtodo numrico. Entonces, el mtodo numrico es estable si existe una constante $C$  independiente de $n$ tal que  
$$E_n \approx C n E_0,$$
en este caso se establece que el error es de crecimiento lineal. Ahora bien, si se encuentra $C>1$ tal que
$$E_n \approx C^n E_0,$$
se dice que el error es de  crecimiento exponencial.

**Definicin** Suponga que $\{\beta_n\}$ es una sucesin cuyo valor de convergencia es cero y que $\{\\alpha_n\}$ es una sucesin que converge a $\alpha$. Si existe una constante positiva $K$ tal que 

$$|\alpha_n - \alpha| \leq K|\beta_n|$$

Entonces decimos que $\{\\alpha_n\}$ converge a $\alpha$ con la rapidz de convergencia de $O\Big(\{\beta_n\}\Big)$. Esta expresin se ller O mayscula de $\{\beta_n\}$. Se indica escribiendo 
$$\alpha_n = \alpha + O\Big(\{\beta_n\}\Big)$$

## Ejemplo 1

Sean las sucesiones $\{a_n\}$ y $\{b_n\}$ definidas por

$$a_n = \frac{n+1}{n^2} \quad \text{y} \quad b_n = \frac{n+3}{n^3}$$

note que ambas sucesiones convergen a cero. Sea $\hat{a}_n=\frac{1}{n}$ y $\hat{b}_n=\frac{1}{n^2}$

$$|a_n-0|=\frac{n+1}{n^2}\leq \frac{n+n}{n^2}=2\frac{1}{n}=2\hat{a}_n$$

y 

$$ |b_n-0|=\frac{n+3}{n^3}\leq \frac{n+3n}{n^3}=4\frac{1}{n^2}=4\hat{b}_n$$

de modo que 

$$a_n=0+O\Big(\frac{1}{n}\Big) \quad \text{y} \quad b_n=0+O\Big(\frac{1}{n^2}\Big)$$

Esto quiere decir que la rpidz de convergencia de $\{b_n\}$ es mayor que la de $\{a_n\}$

```python
import numpy as np
import matplotlib.pyplot as plt
x=np.linspace(0,1,10)
y1=1/x
y2=1/x**2
plt.plot(x,y1,'r--',x,y2,'b--')
plt.xlabel('x')
plt.ylabel('y')
plt.title('velocidad de comparacin')
plt.legend(['y=1/x','y=1/x^2'])
plt.show()
```
