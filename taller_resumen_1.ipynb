6. El siguiente código está programado en python y usa el método de newton 

```python
def newton(f,df,x0,tol=1e-5,maxiter=15):
    x = x0
    for i in range(maxiter):
        dx = -f(x)/df(x)
        x = x + dx
        if abs(dx/x) < tol: return x,i
    print('Son demasiadas iteraciones\n')
```

29. Implemente este código en python

```python

def puntofijo(g, x0, tol, n):
    for k in range(n):
        x1 = g(x0)
        if abs(x1 - x0) < tol:
            return x1, k
        x0 = x1
    print('Son demasiadas iteraciones\n')
g = lambda x: 1 + 1/x
x0 = 1
tol = 1e-5
n = 100
```

```python

def biseccion(f, a, b, tol, n):
    for k in range(n):
        p = a + (b - a)/2
        if abs(f(p)) < tol:
            return p, k
        elif f(a)*f(p) < 0:
            b = p
        else:
            a = p
    print('Son demasiadas iteraciones\n')
f = lambda x: x**2 - 2
a = 0
b = 1
tol = 1e-5
n = 100
```
